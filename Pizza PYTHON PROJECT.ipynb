# Pizza Sales Analysis Project

#### Libraries Used

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Importing csv file using pandas

```python
path= pd.read_csv(r"C:\users\91702\Documents\pizza_sales_excel_file.csv")
```

```python
df= pd.DataFrame(path)
df
```

df.info()

#### Droping  unwanted column


```python
df.drop(columns=["order_time"],inplace=True)
```

#### Converting data types

```python
df["order_date"]=pd.to_datetime(df["order_date"])
df.info()
```

#### Creating Additional Columns as per our Requirement
##### Month_name and Weekday_name

```python
df["Month_name"]= df["order_date"].dt.month_name()
```

```python
df["Weekday_name"]= df["order_date"].dt.day_name()
df
```

#### Replaceing the Data

```python
 df.replace(["L","M","S","XL","XXL"],["Large","Medium","Regular","X-Large","XX-Large"],inplace =True)
```

### KPI CARDS

```python
Total_Revenue = df["total_price"].sum()

Total_Pizza_Sold= df["quantity"].sum()

Total_Orders = df["order_id"].nunique()

Pizza_Size= df["pizza_size"].nunique()

print("TOTAL REVENUE:",f'{Total_Revenue:,.0f}',"\n")
print("TOTAL PIZZA SOLD:",f'{Total_Pizza_Sold:,.0f}',"\n")
print("TOTAL ORDERS:",f'{Total_Orders :,.0f}',"\n")
print("PIZZA SIZES:",f'{ Pizza_Size:,.0f}',"\n")
```

#### Charts 

##### Monthly Wise Total Orders

```python
m_total_orders =df.groupby("Month_name")["order_id"].nunique()

plt.figure(figsize= (28,9))
line=plt.plot(m_total_orders.index,m_total_orders.values,marker ="o",color="darkblue")
plt.fill_between(m_total_orders.index,m_total_orders.values,color="lightpink",alpha=0.2)

plt.xlabel("Months",fontsize= 18)
plt.ylabel("Total Orders",fontsize= 18)
plt.title("MONTHLY WISE TOTAL ORDERS",fontsize= 18)

for x , y in zip(m_total_orders.index,m_total_orders.values):
    plt.text(x,y,f'{y:,.0f}',ha="right",va="bottom",fontsize =18)
    
plt.tight_layout()
plt.show()
```

##### Weekly Total Orders

```python
w_total_orders =df.groupby("Weekday_name")["order_id"].nunique()

plt.figure(figsize=(22,8))
bars= plt.bar(w_total_orders.index,w_total_orders.values,color=["lightyellow","gold"],edgecolor="black",alpha=0.4)


plt.xlabel("Weekday",fontsize= 18)
plt.ylabel("Total Orders",fontsize= 18)
plt.title("WEEKDAY WISE TOTAL ORDERS",fontsize= 18)

for bars in bars:
    plt.text(bars.get_x()+bars.get_width()/2,bars.get_height(),f'{bars.get_height():,.0f}',ha="center",va="bottom",fontsize=15)

plt.tight_layout()
plt.show()
```

#### Total % of Sales By Pizza Size

```python
pizza_sold_category = df.groupby("pizza_size")["total_price"].sum()
pizza_sold_category

plt.figure(figsize = (10,5))
plt.pie(pizza_sold_category,labels=pizza_sold_category.index,colors=["lightgreen","skyblue","lightpink","gold"],autopct= "%.1f%%",shadow=True)

plt.title("Total % of Sales by Pizza Size")

plt.tight_layout()
plt.show()
```

#### Total Sales by Pizza Category

#### Total Quantity Sold by Pizza Category

```python
data=df.groupby("pizza_category")["quantity"].sum().sort_values(ascending =False)

plt.figure(figsize= (18,6))
plt.plot(data.index,data.values,marker='o',linestyle="--",color = "red")

for x , y in zip(data.index,data.values):
    plt.text(x,y,f'{y:,.0f}',ha="left",va="bottom",fontsize =10)

plt.xlabel("Pizza Categroy",fontsize =13)
plt.ylabel("Quantity Sold",fontsize= 13)
plt.title("Total quantity Sold based on Pizza Category",fontsize= 15)

plt.tight_layout()
plt.show()
```

#### Top 6 and Bottom 6 Pizzas By Revenue

```python
x=df.groupby("pizza_name")["total_price"].sum().sort_values(ascending=False).head(6)
y=df.groupby("pizza_name")["total_price"].sum().sort_values(ascending=False).tail(6)

plt.figure(figsize= (20,8))

plt.xticks(rotation= 90)

top6=plt.bar(x.index,x.values,color="orange",edgecolor= "brown",alpha=0.5)
plt.title("Line Chart")
for top6 in top6:
    plt.text(top6.get_x()+top6.get_width()/2,top6.get_height(),f'{top6.get_height():,.0f}',ha="center",va="bottom",fontsize=12)


plt.yticks(rotation= 90)
bottom=plt.bar(y.index,y.values,color="lightgreen",edgecolor= "darkgreen",alpha=0.4)

for bottom in bottom:
    plt.text(bottom.get_x()+bottom.get_width()/2,bottom.get_height(),f'{bottom.get_height():,.0f}',ha="center",va="bottom",fontsize=12)


plt.xlabel("Pizza Names",fontsize= 15)
plt.ylabel("Total Revnue",fontsize= 15)
plt.title("Top and Bottom 6 Pizzas by Revenue")

plt.tight_layout()
plt.show()
```

####  Top 6 Bottom 6 Pizza by Quantity Sold
